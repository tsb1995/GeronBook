```python
import pandas as pd
import numpy as np
import os
```

```python
BLI = pd.read_csv("data\BLI.csv")
WEO = pd.read_csv("data\WEO.xls", thousands=',', delimiter='\t', encoding='latin1', na_values='n/a')
```

```python
WEONew2
```

```python
WEONew2 = WEO.copy().dropna()[['Country', '2020']]
```

```python
BLINew = BLI.copy()
```

```python
BLINew = BLINew[['Country','Value']]
```

```python
BLINew2 = BLI[BLI['INEQUALITY']=='TOT']
BLINew2 = BLINew2.pivot(index="Country", columns="Indicator", values="Value")
```

```python
WEONew2.rename(columns={"2020": "GDP per capita"}, inplace=True)
WEONew2.set_index("Country", inplace=True)
```

```python
full_country_stats = pd.merge(left=BLINew2, right=WEONew2,
                                  left_index=True, right_index=True)
full_country_stats.sort_values(by="GDP per capita", inplace=True)
```
